```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
import time
```

```python
class InstagramBot:
        def __init__(self):
                # Start Chrome browser
                self.driver = webdriver.Chrome()

        def login(self):
            # Open a sample login page
            self.driver.get("https://www.instagram.com/")  # Replace with the actual URL of a login page
    
            # Locate the username and password fields
            username_field = self.driver.find_element(By.NAME, "username")  # Replace "username" with the field's actual ID
            password_field = self.driver.find_element(By.NAME, "password")  # Replace "password" with the field's actual ID

            # Enter credentials
            username_field.send_keys(username)  
            password_field.send_keys(password)  
            time.sleep(2)
            # Submit the form
            password_field.send_keys(Keys.ENTER)

            # Print the current page title (to verify login success)
            print("Logged in to:", self.driver.title)

        def find_user(self,account_name):
            time.sleep(5)
            self.driver.get(f"https://www.instagram.com/{account_name}")
            time.sleep(2)
            

            try:
                follow = self.driver.find_element(By.XPATH, "//button[normalize-space()='Follow']")

                follow.click()
                print(f"Followed To {account_name}")
            
            except:
                print(f"Already following {account_name} ")
            
            followers = self.driver.find_element(By.XPATH, "//a[@href='/virat.kohli/followers/']//span[contains(@class, 'xdj266r')]")
            followers.click()
        
       
        def follow(self, number):
            follow_buttons = []

            for i in range(1, number + 1):
                try:
                    # Use a relative XPath to find follow buttons
                    btn = WebDriverWait(self.driver,20).until(
                        EC.presence_of_element_located((By.XPATH, f"(//button[.//div[contains(text(), 'Follow')]])[{i}]"))
                    )
                    follow_buttons.append(btn)
                except NoSuchElementException:
                    print(f"Button not found at index {i}")
                    continue

            for button in follow_buttons:

                try:
                    button.click()
                    time.sleep(1)
                except ElementClickInterceptedException:
                    # Close modal or dismiss any interfering element
                    try:
                        modal_close = self.driver.find_element(By.XPATH, "//button[contains(@class, 'dismiss')]")
                        modal_close.click()
                        print("Closed interfering modal.")
                        button.click()  # Retry clicking
                    except:
                        print("No modal found or couldn't close it.")


        def close(self):
            self.driver.quit()
```
